### Frame Extraction based on Button Delay

first extract duration from video file with ffmpeg

```python
import subprocess
import datetime

duration_list = []

for i in df.index:
    x,y = 0,0
    gazeX = df[' gazeX'].values[i]
    gazeY = df[' gazeY'].values[i]
    video_num = df['VideoNum'].values[i]
    out = subprocess.run(["ffprobe", files[i]], capture_output=True)
    duration = str(out).split("Duration: ")[-1].split(',')[0]
    if int(gazeY) <= 392:
        y=0
    elif int(gazeY) <= 752:
        y=1
    elif int(gazeY) <= 932:
        y=2
    elif int(gazeY) <= 1112:
        y=3
    elif int(gazeY) <= 1832:
        y=4
    elif int(gazeY) <= 2192:
        y=5
    else:
        y=6

    if int(gazeX) <= 303.5:
        x=1
    elif int(gazeX) <= 580.5:
        x=2
    elif int(gazeX) <= 857.5:
        x=3
    elif int(gazeX) <= 1134.5:
        x=4
    else:
        x=5

    button_num = y*5+x
    duration_list.append([button_num, video_num,gazeX,gazeY,duration,files[i]])
duration_df = pd.DataFrame(duration_list,columns=['button_num', 'video_num','gazeX', 'gazeY','duration','video_name'])
duration_df
```

```python
import numpy as np
import cv2

def improve_contrast_image_using_clahe(bgr_image: np.array) -> np.array:
    hsv = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2HSV)
    hsv_planes = cv2.split(hsv)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    hsv_planes[2] = clahe.apply(hsv_planes[2])
    hsv = cv2.merge(hsv_planes)
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
```

```python
images = glob.glob("/special/GazeVideo/CaptureApp/image/one*")
images.sort()
img = cv2.imread(images[0])
img = improve_contrast_image_using_clahe(img)
img = Image.fromarray(img)
img
```
